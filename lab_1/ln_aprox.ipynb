# Logarithm aproximation

The following code aproximates the logarithm function using an itertative method as described in:
    
Carlson, B. C. (1972). An algorithm for computing logarithms and arctangents. Mathematics of Computation, 26(118), 543–549. https://doi.org/10.1090/s0025-5718-1972-0307438-2

The method is based on computing the arithmetic and geometric mean of two values $a_i$ and $g_i$ in the following way:

*  With a given value $x>0$, we initialize $a_0 = \frac{(1+x)}{2}$, $g_0=\sqrt{x}$;
*   We iterate $a_{i+1}=\frac{a_i+g_i}{2}$ and $g_{i+1}=\sqrt{a_{i+1}g_i}$;
*   $\frac{x-1}{a_i}$ can be then considered an approximation to $\ln(x)$

Task 1: The following function aproximate the ln function by n steps using the iterative method cited above.

```python
import numpy as np
import matplotlib.pyplot as plt

def approx_ln(x,n):
    '''
    The function takes in an x value and returns the aproximated ln(x) ≈ (x-1)/a_n
    after n iterations, as described in https://doi.org/10.1090/s0025-5718-1972-0307438-2
    '''
    #initializing starting values
    a0 = (1 + x)/2
    g0 = np.sqrt(x)

    if x>0:
        for i in range(n+1):
            ai = (a0 + g0)/2
            gi= np.sqrt(ai * g0)

            a0 = ai
            g0 = gi
    else:
        raise Exception(f'{x} is not greater than x, the ln(x) is not defined')

    return (x-1)/a0
```

Task 2: We then plotted the numpy.ln and our approx_ln for n iterations.


```python
x_asse = np.linspace(0.1, 50, 100) #starting at 0.1 because ln(0) is not defined
n = 7

#the approx_ln works with scalars so it needs to be vectorized to use it with an array
vec_approx_ln = np.vectorize(approx_ln)


fig, ax = plt.subplots()
plt.title('ln(x) and approx_ln for n iterations')

for i in range(2,n+1):
    ax.plot(x_asse, vec_approx_ln(x_asse, i), label=f'n = {i}')




ax.plot(x_asse,np.log(x_asse), 'k--', label='ln(x)')



#zoomed in portion
x1 = x_asse[90:]

zoom = ax.inset_axes([0.55,0.1,0.4,0.4]) #loc and size

for i in range(2,n+1):
    zoom.plot(x1,vec_approx_ln(x_asse, i)[90:])

zoom.plot(x1, np.log(x_asse)[90:], 'k--')
ax.indicate_inset_zoom(zoom)

plt.xlabel('x')
plt.ylabel('ln(x)')
plt.legend(loc=2)
plt.show()
```

Task 3: We plotted the difference between the numpy.ln and the approximation

```python
plt.figure()
for i in range(2,n+1):
    plt.plot(x_asse, np.log(x_asse)-vec_approx_ln(x_asse, i), label=f'n = {i}')


plt.title('Differences between ln(x) and approx_ln for n iterations')
plt.xlabel('x')
plt.xlim(1)
plt.ylabel('ln(x) - aprx_ln(x)')
plt.legend()
plt.show()
```

Task 4: Considering x = 1.41 we plotted the absolute value of the error versus the number of iterations n.

```python
x = 1.41
error = []
for i in range(2,n+1):
    error.append(np.abs(np.log(x)-approx_ln(x, i)))


plt.plot(np.linspace(2,n,6), error, label='ln(1.41) - aprx_ln(1.41)')
plt.title('Absolute value of the error vs number of interations')
plt.xlabel('n')
plt.ylabel('abs error')
plt.legend()
plt.show()
```

Task 5: In the article cited above a method is suggested to accelerate the convergence and it involves for $i=0,...,n$:
*   $d_{0,i}=a_i$
*   $d_{k,i}=\frac{d_{k-1,i}-4^{-k}d_{k-1,i-1} }{1-4^{-k}}$ with $k=1,...,i$ for $i>0$

This new faster method then takes as an approximation for the logarithm the value $\frac{x-1}{d_{n,n}}$

The following function implements that approach

```python
def fast_approx_ln(x,n):
    '''
    The function is the accelerated Carlsson method to approximate the log.
    It takes a x value and returns the approximated ln(x) ≈ (x-1)/d_n_n
    after n iterations, as described in https://doi.org/10.1090/s0025-5718-1972-0307438-2
    '''
    #initializing starting values
    a0 = (1 + x)/2
    g0 = np.sqrt(x)

    #and a list to store d[i][k]
    d = []
    #it's a list of lists where:
    #   d[i][0] is the apprx we'd get from approx_ln after i iterations
    #   d[i][k] is the value at i-iteration and k-level

    for i in range(n+1):
        ai = (a0 + g0)/2
        gi= np.sqrt(ai * g0)

        a0 = ai
        g0 = gi

        #initial d[i][0] value
        d.append([ai])

        for k in range(1,i+1):
            dik = (d[i][k-1] - 4**(-k) * d[i-1][k-1]) / (1 - 4**(-k))

            d[i].append(dik)

    return (x-1)/d[n][n]
```

Task 6: the folllowing is an error plot for the faster aprx

```python
x_asse = np.linspace(0.1, 20, 1000)

vec_fast_approx_ln = np.vectorize(fast_approx_ln)


plt.figure()


for i in range(2,n):
    error = np.abs(np.log(x_asse) - vec_fast_approx_ln(x_asse, i))

    plt.scatter(x_asse, error, label=f'Iteration {i}', s=10)



plt.title('Error behavior of the accelerated Carlsson method for the log')
plt.xlabel('x')
plt.ylabel('abs error')
plt.yscale('log') #plotting the y axis in log scale to match the given plot
plt.ylim(1e-19,1e-5)
plt.xlim(0.0,20.0)
plt.legend()


plt.show()
```
